```python
from explainer import ExplainableSTS

# initialise the explainable sts model
model = ExplainableSTS()
```

```python
# entity swap
s1 = 'Charlton coach Guy Luzon had said on Monday'
s2 = 'Charlton coach Bordeaux had said on Monday'

# get the similarity score between the sentences
sim_score = model(s1, s2) 
print('predicted similarity score:', sim_score)

# get the explanation
model.explain(s1, s2, plot=True)
```

```python
# negation
s1 = 'Snow was predicted later in the weekend for Atlanta and areas even further south.'
s2 = 'Snow wasnt predicted later in the weekend for Atlanta and areas even further south.'

sim_score = model(s1, s2) 
print('predicted similarity score:', sim_score)
model.explain(s1, s2, plot=True)
```

```python
# pronoun swap
s1 = 'Tom is his father'
s2 = 'Tom is her dad'

sim_score = model(s1, s2) 
print('predicted similarity score:', sim_score)
model.explain(s1, s2, plot=True)
```

```python
# number change
s1 = 'He says he wants to pay off the $12.6 million lien'
s2 = 'He says he wants to pay off the $3.45 million lien'

sim_score = model(s1, s2) 
print('predicted similarity score:', sim_score)
model.explain(s1, s2, plot=True)
```

```python
# proposition change, good
s1 = 'Her birthday was in July'
s2 = 'Her birthday was before July'

sim_score = model(s1, s2) 
print('predicted similarity score:', sim_score)
model.explain(s1, s2, plot=True)
```
